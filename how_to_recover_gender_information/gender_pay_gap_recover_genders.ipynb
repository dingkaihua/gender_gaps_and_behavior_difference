```python
name_list
```

```python
first_names=[]

for name in name_list:
  #print(name)
  name_lower = name.lower()
  name_parts = name_lower.split(',')
  #print(name_parts)
  if len(name_parts[-1]) != 0:
    first_name = name_parts[-1]
  else:
    first_name = name_parts[0]
  first_name_string = first_name.split(' ')

  # This table contains 2 naming conventions,
  # (1) family name, middle name initial + last name
  # (2) family name, first name + middle name
  if (len(first_name_string) !=1) and (len(first_name_string[1])!=1):
    concise_first_name = first_name_string[1]
  elif (len(first_name_string) !=1) and (len(first_name_string[1])==1):
    concise_first_name = first_name_string[-1]
  else:
    concise_first_name = first_name_string[0]
  #if concise_first_name.capitalize() == 'P':
   # break
  #print(concise_first_name)
  first_names.append(concise_first_name.capitalize())
```

```python
print(first_names)
```

```python
first_names_df = pd.DataFrame(first_names)
```

```python
import gender_guesser.detector as gender

genders = []
detector = gender.Detector()

for name in first_names:
  detected_gender = detector.get_gender(name)
  # post processing the gender

  if detected_gender == 'mostly_female':
    detected_gender = 'female'
  elif detected_gender == 'mostly_male':
    detected_gender = 'male'


  genders.append(detected_gender)
```
